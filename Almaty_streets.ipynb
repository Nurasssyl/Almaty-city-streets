```python
import osmnx as ox
```

```python
admin = ox.geocode_to_gdf('Astana')
admin_poly = admin.geometry.values[0]
admin_poly
```

```python
G = ox.graph_from_polygon(admin_poly)
```

```python
G.number_of_nodes(), G.number_of_edges
```

```python
nodes, edges = ox.graph_to_gdfs(G)
```

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
f, ax = plt.subplots(1, figsize=(10, 10))
edges.plot(ax=ax, edgecolor = 'orange', linewidth = 0.3)
ax.axis('off')
```

```python
# Сохранение визуализации
output_path = 'almaty_street_network.png'
f.savefig(output_path, dpi=300, bbox_inches='tight', facecolor=f.get_facecolor())

from google.colab import files
files.download(output_path)
```
